```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("data/Global YouTube Statistics.csv",encoding='unicode escape')
```

```python
df.head()
```

```python
df.columns
```

```python
youtube_df=df[['rank', 'Youtuber','category','Country','date_created', 'subscribers','subscribers_for_last_30_days',
               'video views','video_views_for_the_last_30_days', 'channel_type', 'channel_type_rank',
               'video_views_rank', 'country_rank','lowest_monthly_earnings',
             'highest_monthly_earnings', 'lowest_yearly_earnings','highest_yearly_earnings', 
       'Population','Urban_population', 'Latitude', 'Longitude']]
```

```python
youtube_df=youtube_df.rename(columns={'rank':'id', 'Youtuber':'youtube_name',  'video views':'video_views',\
                                        'Country':'country','Population':'population','Urban_population':'urban_population',\
                                        'Latitude':'latitude', 'Longitude':'longitude'})
youtube_df.columns
```

```python
youtube_df.shape
```

```python
youtube_df=youtube_df.dropna(how='any')
```

```python
youtube_df.dtypes
```

```python
youtube_df.head()
```

```python
youtube_df['id'] = range(len(youtube_df))

youtube_df.head()
```

```python
youtube_df.shape
```

```python
youtube_df.to_json('data/youtube.json', orient = 'split', compression = 'infer', index = 'true')
```

```python
conn = sqlite3.connect('data/youtube_db.db')

cursor = conn.cursor()

youtube_df.to_sql('youtube', conn, if_exists='replace', index=False)

# Create a new table with a primary key constraint

cursor.execute("CREATE TABLE youtube_table (id INTEGER PRIMARY KEY,\
                youtube_name VARCHAR,\
                category  VARCHAR,\
                country   VARCHAR,\
                date_created  VARCHAR,\
                subscribers   REAL,\
                subscribers_for_last_30_days  REAL,\
                video_views  REAL,\
                video_views_for_the_last_30_days REAL,\
                channel_type  VARCHAR,\
                channel_type_rank REAL,\
                video_views_rank  REAL,\
                country_rank  REAL,\
                lowest_monthly_earnings  REAL,\
                highest_monthly_earnings REAL,\
                lowest_yearly_earnings  REAL,\
                highest_yearly_earnings REAL,\
                population  REAL,\
                urban_population REAL,\
                latitude  REAL,\
                longitude REAL)")
         

cursor.execute("INSERT INTO youtube_table SELECT * FROM youtube")

conn.commit()             
# Close the database connection
conn.close()
```
